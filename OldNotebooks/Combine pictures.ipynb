```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#OCTOPI
fov_oct=1.6*10**-3

#high mag
pixel_size=0.365623*10**-6
nb_pixel_x=3047
nb_pixel_y=3047

fov_nh_x=nb_pixel_x*pixel_size#nh = nighthawk, i.e. high mag
fov_nh_y=nb_pixel_y*pixel_size
```

```python
Folder_highMag='../Images High Mag/'
Folder_lowMag='../Images Low Mag/'
Folder_concatenated='../Images_concatenated/'
```

# looping through different images, flip and rotate 

```python
img=cv2.imread(Folder_highMag+'3B_scan2 - Stitched - attemp2_001_001.tif',0)
```

```python
img=img.transpose()
img=cv2.flip(img,0)
img=cv2.flip(img,1)
```

```python
img=img*4
```

```python
plt.imshow(img,cmap='gray', vmin=0, vmax=255)

```

```python
import os

for filename in os.listdir(Folder_highMag):
    if filename.endswith(".tif"): 
        print(filename)
        img=cv2.imread(Folder_highMag+filename,0)
        img=img.transpose()
        img=cv2.flip(img,0)
        img=cv2.flip(img,1)
        nbrs=filename[-11:-4]
        nb1=int(nbrs[0:3])
        nb2=int(nbrs[4:8])
        new_name=str(nb1)+'_'+str(nb2)+'.tif'
        cv2.imwrite('../HighMag_RF/'+new_name,img)
    else:
        continue
```

# Get the 9 frames for a given picture

```python
import os

for filename in os.listdir(Folder_lowMag):
    if 'fluorescent' in filename:
        print(filename)
        low_mag_id_1=int(filename[1:3])
        low_mag_id_2=int(filename[4:6])
        y_start=max(low_mag_id_1*fov_oct-10**-4,0)
        x_start=max(low_mag_id_2*fov_oct-10**-4,0)
        y_end=y_start+fov_oct+10**-4
        x_end=x_start+fov_oct+10**-4
        
        #loop throughthe frames
        #identify which ones are the limit ones
        nb_im_x=18
        nb_im_y=18

        for i in range(nb_im_x):
            if i*fov_nh_x<=x_start and (i+1)*fov_nh_x>=x_start:
                frame_x_start=i+1
            if i*fov_nh_x<=x_end and (i+1)*fov_nh_x>=x_end:
                frame_x_end=i+1
            if i*fov_nh_y<=y_start and (i+1)*fov_nh_y>=y_start:
                frame_y_start=i+1
            if i*fov_nh_y<=y_end and (i+1)*fov_nh_y>=y_end:
                frame_y_end=i+1
        
        print(frame_x_start,frame_x_end,frame_y_start,frame_y_end)
        #create blank picture
        full_image=np.zeros((nb_pixel_x*(frame_x_end-frame_x_start+1),nb_pixel_y*(frame_y_end-frame_y_start+1)))
        
        #pad picture with corresponding frames
        idx_x=np.arange(frame_x_end,frame_x_start-1,-1)
        idx_y=np.arange(frame_y_end,frame_y_start-1,-1)

        for i in range(idx_x.shape[0]):
            for j in range(idx_y.shape[0]):
                filename=str(idx_x[i])+'_'+str(idx_y[j])+'.tif'
                crt_img=cv2.imread('../HighMag_RF/'+filename,0)
                full_image[i*nb_pixel_x:(i+1)*nb_pixel_x,j*nb_pixel_y:(j+1)*nb_pixel_y]=crt_img
        
        full_image_name=str(low_mag_id_1)+'_'+str(low_mag_id_2)+str('.png')
        full_image=full_image*4
        cv2.imwrite(Folder_concatenated+full_image_name,full_image)
        
    else:
        continue
```
