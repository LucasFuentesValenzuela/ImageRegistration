```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import re
import os

import tifffile as tiff
```

```python
#OCTOPI
fov_oct=1.6*10**-3

#high mag
pixel_size=0.365623*10**-6
nb_pixel_y=3081
nb_pixel_x=2755

fov_nh_x=nb_pixel_x*pixel_size#nh = nighthawk, i.e. high mag
fov_nh_y=nb_pixel_y*pixel_size

Folder_highMag='../HighMag_0/'
Folder_lowMag='../Images Low Mag/'
Folder_concatenated='../Images_concatenated/'
Folder_HighMag_RF='../HighMag_0_RF/'
Folder_Virtual='../Virtual_0/'
```

# looping through different images, flip and rotate 

```python
for filename in os.listdir(Folder_highMag):
    if filename.endswith(".tif"): 
        print(filename)
        #read only the first channel, which corresponds to DAPI
        img=tiff.imread(Folder_highMag+filename)
        img=img[1,:,:]
        img=img.transpose()
        img=cv2.flip(img,0)
        img=cv2.flip(img,1)
        nbrs=filename[-11:-4]
        nb1=int(nbrs[0:3])
        nb2=int(nbrs[4:8])
        new_name=str(nb1)+'_'+str(nb2)+'.tif'
        cv2.imwrite(Folder_HighMag_RF+new_name,img)
    else:
        continue
```

# Build Virtual Image

```python
size_cropping=2.2*10**-3

n_pixels_map=round(size_cropping/pixel_size)
```

```python
delta=(size_cropping-fov_oct)/2
```

```python
#there is an offset in the different picture
offset_x=3000*pixel_size
offset_y=1500*pixel_size
```

There is without a doubt a way to speed this up by pasting directly slices of the matrices into the new one

```python
#With "virtual image" -- pixel by pixel

for filename in os.listdir(Folder_lowMag):
    if 'fluorescent' in filename:
        
        print(filename)
        low_mag_id_1=int(filename[1:3])
        low_mag_id_2=int(filename[4:6])
        y_start=max(low_mag_id_1*fov_oct-delta,0)+offset_y
        x_start=max(low_mag_id_2*fov_oct-delta,0)+offset_x
#         y_end=max(low_mag_id_1*fov_oct,0)+fov_oct+2*10**-4+offset_y
#         x_end=max(low_mag_id_2*fov_oct,0)+fov_oct+2*10**-4+offset_x
        y_end=y_start+fov_oct+2*delta
        x_end=x_start+fov_oct+2*delta
        
        #loop through the frames
        #identify which ones are the limit ones
        nb_im_x=18

        for i in range(nb_im_x):
            if i*fov_nh_x<=x_start and (i+1)*fov_nh_x>=x_start:
                frame_x_start=i+1
            if i*fov_nh_x<=x_end and (i+1)*fov_nh_x>=x_end:
                frame_x_end=i+1
            if i*fov_nh_y<=y_start and (i+1)*fov_nh_y>=y_start:
                frame_y_start=i+1
            if i*fov_nh_y<=y_end and (i+1)*fov_nh_y>=y_end:
                frame_y_end=i+1
        
        print(frame_y_start,frame_y_end, '--',frame_x_start,frame_x_end)
        #create blank picture
        full_image=np.zeros((n_pixels_map,n_pixels_map))
        #pad picture with corresponding frames
        idx_x=np.arange(frame_x_end,frame_x_start-1,-1)
        idx_y=np.arange(frame_y_end,frame_y_start-1,-1)

        for i in range(idx_x.shape[0]):
            for j in range(idx_y.shape[0]):
                filename=str(idx_y[j])+'_'+str(idx_x[i])+'.tif'
                if os.path.exists(Folder_HighMag_RF+filename):
                    crt_img=cv2.imread(Folder_HighMag_RF+filename,0)
                else:
                    continue
                
                #limits of current image
                x_start_crt=nb_pixel_x*(idx_x[i]-1)*pixel_size
                y_start_crt=nb_pixel_y*(idx_y[j]-1)*pixel_size
                x_end_crt=x_start_crt+nb_pixel_x*pixel_size
                y_end_crt=y_start_crt+nb_pixel_y*pixel_size
                
                print("TEST:",idx_x[i],idx_y[j])
                
                #find the indices that correspond to the mapping
                
                
                ## We have to speed up this part of the code
                ## This works but is far too slow
                for n in range(n_pixels_map):
                    for m in range(n_pixels_map):
                        #actual positions in the new map
                        x_=x_end-n*pixel_size
                        y_=y_end-m*pixel_size
                        
                        if x_>=x_start_crt and x_<= x_end_crt and y_>=y_start_crt and y_<=y_end_crt:
                            #indices
                            x_idx=int(round((x_end_crt-x_)/pixel_size))
                            y_idx=int(round((y_end_crt-y_)/pixel_size))
                            if x_idx > nb_pixel_x-1 or y_idx > nb_pixel_y-1 or x_idx < 0 or y_idx <0:
                                continue
                            else:
                                full_image[m,n]=crt_img[y_idx,x_idx]

        
        full_image_name=str(low_mag_id_1)+'_'+str(low_mag_id_2)+str('.png')
        cv2.imwrite(Folder_Virtual+full_image_name,full_image)
        print(full_image_name)
        print(full_image.shape)
        print("---------------")
        
    else:
        continue
```

```python
#with concatenation, too

for filename in os.listdir(Folder_lowMag):
    if 'fluorescent' in filename:
        
        print(filename)
        low_mag_id_1=int(filename[1:3])
        low_mag_id_2=int(filename[4:6])
        y_start=max(low_mag_id_1*fov_oct-delta,0)+offset_y
        x_start=max(low_mag_id_2*fov_oct-delta,0)+offset_x
#         y_end=max(low_mag_id_1*fov_oct,0)+fov_oct+2*10**-4+offset_y
#         x_end=max(low_mag_id_2*fov_oct,0)+fov_oct+2*10**-4+offset_x
        y_end=y_start+fov_oct+2*delta
        x_end=x_start+fov_oct+2*delta
        
        #loop through the frames
        #identify which ones are the limit ones
        nb_im_x=18

        for i in range(nb_im_x):
            if i*fov_nh_x<=x_start and (i+1)*fov_nh_x>=x_start:
                frame_x_start=i+1
            if i*fov_nh_x<=x_end and (i+1)*fov_nh_x>=x_end:
                frame_x_end=i+1
            if i*fov_nh_y<=y_start and (i+1)*fov_nh_y>=y_start:
                frame_y_start=i+1
            if i*fov_nh_y<=y_end and (i+1)*fov_nh_y>=y_end:
                frame_y_end=i+1
        
        print(frame_y_start,frame_y_end, '--',frame_x_start,frame_x_end)
        #create blank picture
#         full_image=np.zeros((n_pixels_map,n_pixels_map))
        full_image=np.zeros((nb_pixel_y*(frame_y_end-frame_y_start+1),nb_pixel_x*(frame_x_end-frame_x_start+1)))
        #pad picture with corresponding frames
        idx_x=np.arange(frame_x_end,frame_x_start-1,-1)
        idx_y=np.arange(frame_y_end,frame_y_start-1,-1)

        for i in range(idx_x.shape[0]):
            for j in range(idx_y.shape[0]):
                filename=str(idx_y[j])+'_'+str(idx_x[i])+'.tif'
                if os.path.exists(Folder_HighMag_RF+filename):
                    crt_img=cv2.imread(Folder_HighMag_RF+filename,0)
                    print(crt_img.shape)
                else:
                    continue
                full_image[j*nb_pixel_y:(j+1)*nb_pixel_y,i*nb_pixel_x:(i+1)*nb_pixel_x]=crt_img
                
                if i==0 and j==0:
                    x_start_crt=nb_pixel_x*(idx_x[i]-1)*pixel_size
                    y_start_crt=nb_pixel_y*(idx_y[j]-1)*pixel_size
                    x_ref=x_start_crt+nb_pixel_x*pixel_size
                    y_ref=y_start_crt+nb_pixel_y*pixel_size
                    
        x_idx=int(round((x_ref-x_end)/pixel_size))
        y_idx=int(round((y_ref-y_end)/pixel_size))
        
        full_image=full_image[x_idx:x_idx+n_pixels_map,y_idx:y_idx+n_pixels_map]
                
        
        full_image_name=str(low_mag_id_1)+'_'+str(low_mag_id_2)+str('.png')
        cv2.imwrite(Folder_Virtual+full_image_name,full_image)
        print(full_image_name)
        print(full_image.shape)
        print("---------------")
    else:
        continue
```
